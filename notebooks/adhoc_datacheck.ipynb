```python
import pandas as pd 
from airflow_practice.config import settings
```

```python
def read_tables(): 
    for path in settings.data_path.glob(".raw/*"): 
        yield pd.read_table(path, delimiter="\s+", header=None, dtype="str")
        
```

```python
df = pd.concat(read_tables())
df.head(2)
```
